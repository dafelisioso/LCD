```python
# dataframe opertations - pandas
import pandas as pd
# plotting data - matplotlib
from matplotlib import pyplot as plt
# time series - statsmodels 
# Seasonality decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.seasonal import seasonal_decompose 
# holt winters 
# single exponential smoothing
from statsmodels.tsa.holtwinters import SimpleExpSmoothing   
# double and triple exponential smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
m = 12
alpha = 1/(2*m)
```

```python
df['HWES1'] = SimpleExpSmoothing(df['Precio / Utilidad x']).fit(smoothing_level=alpha,optimized=False,use_brute=True).fittedvalues
df[['Precio / Utilidad x','HWES1']].plot(title='Holt Winters Single Exponential Smoothing');
```

```python
df['HWES2_add'] = ExponentialSmoothing(df['Precio / Utilidad x'],trend='add').fit().fittedvalues
df['HWES2_mul'] = ExponentialSmoothing(df['Precio / Utilidad x'],trend='mul').fit().fittedvalues
df[['Precio / Utilidad x','HWES2_add','HWES2_mul']].plot(title='Holt Winters Double Exponential Smoothing: Additive and Multiplicative Trend');
```

```python
df['HWES3_ADD'] = ExponentialSmoothing(df['Precio / Utilidad x'],trend='add',seasonal='add',seasonal_periods=12).fit().fittedvalues
df['HWES3_MUL'] = ExponentialSmoothing(df['Precio / Utilidad x'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
df[['Precio / Utilidad x','HWES3_ADD','HWES3_MUL']].plot(title='Holt Winters Triple Exponential Smoothing: Additive and Multiplicative Seasonality');
```

```python
df.index.freq = 'MS'
train_airline=df
test_airline=df
```

```python
fitted_model = ExponentialSmoothing(df['Precio / Utilidad x'],trend='mul',seasonal='mul',seasonal_periods=12).fit()
test_predictions = fitted_model.forecast(24)
train_airline['Precio / Utilidad x'].plot(legend=True,label='TRAIN')
test_airline['Precio / Utilidad x'].plot(legend=True,label='TEST')
test_predictions.plot(legend=True,label='PREDICTION')
plt.title('Train, Test and Predicted Test using Holt Winters')
```
