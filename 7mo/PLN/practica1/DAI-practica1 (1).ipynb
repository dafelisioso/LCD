```python
#importamos las librerias y cargamos nuestros datos
import re
from dateutil import parser

from dateutil import parser
with open("dates.txt", "r") as myfile:
  string=myfile.read().replace('\n','')
```

```python
# Buscamos las expresiones regulares con 
# dia -mes -año (número)
l1 = re.compile(r'[\d]{1,2}\/[\d]{1,2}\/[\d]{2,4}|[\d]{1,2}\-[\d]{1,2}\-[\d]{2,4}')
fechas = l1.findall(string)
fechas.remove('9/36/308')

print(fechas)
print(len(fechas))

#las ordenamos por fecha
fec=list()
for i in fechas:
    d = parser.parse(i)
    fec.append(d.strftime("%Y-%m-%d "))
    


fecha1=sorted(fec)

fecha1.remove('9712-08-09 ')

fecha1.remove('2071-09-12 ')

fecha1.remove('2071-07-11 ')

fecha1.remove('2071-05-18 ')

fecha1.remove('2071-07-08 ')
fecha1.pop(-1)
print(fecha1)
print(len(fecha1))
```

```python
# mes día año (letra)
l2 = re.compile(r'[JjFfMmAaJjSsOcNnDe][a-z,A_Z]+\.? [\d]{1,2}, [\d]{2,4}')
fechas2 = l2.findall(string)
fechas2.pop(-2)
fechas2.pop(-1)
print(fechas2)
print(len(fechas2))
fec2=list()
for i in fechas2:
    d = parser.parse(i)
    fec2.append(d.strftime("%Y-%m-%d "))
    


fechas2=sorted(fec2)
print(fechas2)
print(len(fechas2))
```
