```python
import numpy as np

class Perceptron(object):
  def __init__(self,eta=0.01,n_iter=50,random_state=1):
    """ docstring del consructor"""
    self.eta=eta
    self.n_iter=n_iter
    self.random_state=random_state
  
  def fit(self,X,y):
    """ajuste=aprendizaje"""
    rgen =np.random.RandomState(self.random_state)
    if self.random_state is None:
      self.w_=np.zeros(1+len(X[0]))
    else:
      self.w_ = rgen.normal(loc=.0,scale=0.1,size=1+X.shape[1])
    self.errors_ =[]
    for _ in range(self.n_iter):
      errors =0
      for xi ,yi in zip(X,y):
        update= self.eta *(yi-self.predict(xi))
        self.w_[1:] += update  * xi
        self.w_[0] += update

        errors += int(update != 0.0)

      self.errors_.append(errors)
    return self

  def net_input(self,X):
    return np.dot(X,self.w_[1:]) + self.w_[0]
  def predict(self,X):
    return np.where(self.net_input(X) >= 0.0 ,1,-1)
```

```python
X=np.array([[0,0],[0,1],[1,0],[1,1],[1,2]])
y = np.array([-1,1,1,1,-1])
ppn = Perceptron(n_iter=7,eta=0.1,random_state=None)
ppn.fit(X,y)
print("Pesos:", ppn.w_)
```

```python
cov_mat=np.cov(X_train_std.T)
eigen_vals,eigen_vecs=np.linalg.eig(cov_mat)
print('Eigenvals:',eigen_vals)
print('eigenvecs:',eigen_vecs)
```

```python
import matplotlib.pyplot as plt
tot= sum(eigen_vals)
var_exp=[ev/tot for ev in sorted(eigen_vals,reverse= True)]
cum_var_exp =np.cumsum(var_exp)
plt.bar(range(1,14),var_exp,label='varianza individual',align='center')
plt.step(range(1,14),cum_var_exp,where='mid',label='varianza acumulativa')
plt.xlabel('componenetes')
plt.ylabel('varianza')
plt.legend(loc='best')
plt.show()
```

```python
#KPCA con RBF

import scipy
from scipy.spatial.distance  import pdist, squareform
from scipy.linalg import eigh
import numpy as  np
import matplotlib.pyplot as plt
```

```python
def rbf_kpca(X,gamma,n_componenets):
  sq_dist=pdist(X,'sqeuclidean')
  mat_sq_dist= squareform(sq_dist)
  K= np.exp(-gamma*mat_sq_dist)
  N= K.shape[0]
  one_n=np.ones((N,N))/N
  
  K =K-one_n.dot(K) -K.dot(one_n)+one_n.dot(K).dot(one_n)
  eigen_vals,eigen_vecs= eigh(K)
  eigen_vals,eigen_vecs=eigen_vals[::-1],eigen_vecs[:,::-1]
  print(eigen_vals.shape, ':',eigen_vecs.shape)

#selaccionar K
  alphas =np.column_stack([eigen_vecs[:,i] for i in range(n_componenets)])
  lambdas = np.column_stack([eigen_vals[i] for i in range(n_componenets)])

  return alphas,lambdas
```

```python
data=np.array([[8, 10], [3, 10.5], [7, 13.5], [5, 18], [5, 13], [6, 9], [9, 11], [3, 18], [8.5, 12], [8, 16]])
C=np.array([[8,10],[3,10.5]])
fig=plt.figure()
plt.scatter(C[0][0],C[0][1],marker='*',s=200,c='k')
plt.scatter(C[1][0],C[1][1],marker='*',s=200,c='k')
plt.scatter(data[:,0],data[:,1])
```

```python
distances=[]
clusters=np.zeros(len(data))
def dist(a,b,ax=1):
  return np.linalg.norm(a-b,axis=ax)

for i in range(len(data)):
  distance=dist(data[i],C)
  distances.append(distance)
  cluster=np.argmin(distance)
  clusters[i]= cluster
print(clusters)
distances
```

```python
for i in range(len(C)):
  points=[data[j] for j in range(len(data)) if clusters[j]==i]
  C[i]=np.mean(points,axis=0)
C
```

```python
fig=plt.figure()
plt.scatter(C[0][0],C[0][1],marker='*',s=200,c='k')
plt.scatter(C[1][0],C[1][1],marker='*',s=200,c='k')
plt.scatter(data[:,0],data[:,1])
```

```python
#metricas
from sklearn.metrics import mean_absolute_error

def mean_absolute_porcentage_error(y_true,y_pred):
  return np.mean(np.abs((y_true-y_pred)/y_true))*100
```

```python
#media movil
def moving_ave(serie,n):
  ##calcula ma media mvil de una serie de tiempo
  #recibe: serie:np.array con los datos de la serie
  # n : el retraso de la ventana para calcular la media mvil
  #regresa: un escalar con los datos de la media para la ventrana indicada
  return np.average(serie[-n:])
moving_ave(ads,24)
```

```python
def plot_MA(serie, window,plot_intervals=False,scale=1.96,plot_anomlies=False):
  rolling_mean = serie.rolling(window=window).mean()
  plt.figure(figsize=(15,5))
  plt.title('Media mvil \n tamao de ventana={}'.format(window))
  plt.plot(rolling_mean,'g', label='Tendencia')
  if plot_intervals:
    mae=mean_absolute_error(serie[window:],rolling_mean[window:])
    dev=np.std(serie[window:]-rolling_mean[window:])
    upper_ =rolling_mean-(mae+scale*dev)
    lower_ =rolling_mean+(mae+scale*dev)
    plt.plot(upper_,'r--',label='limites superior/inferior')
    plt.plot(lower_,'r--')
    if plot_anomlies:
      anomalies =pd.DataFrame(index=serie.index,columns=serie.columns)
      anomalies[serie < lower_] =serie[serie < lower_]
      anomalies[serie > upper_] =serie[serie > lower_]
      plt.plot(anomalies, 'ro',markersize=10)

  plt.plot(serie[window:],label='valores reales')
  plt.legend(loc='upper left')
  plt.grid(True)
```

```python
def weighted_MA(serie,weights):
  result=0.0
  weights.reverse()
  for n in range(len(weights)):
    result += serie.iloc[-n-1]+weights[n]
  return float(result)
```
